```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# File to Load (Remember to change these)
all_games_to_load = "Data/all_games.csv"
video_games_to_load = "Data/video_games.csv"

# Read the all games and video games Data
all_games_df = pd.read_csv(all_games_to_load)
video_games_df = pd.read_csv(video_games_to_load)
```

```python
# Rename Name to name column in video_games_df
video_games_df.columns = video_games_df.columns.str.lower()
video_games_df
```

```python
# renaming platform types abrv in video_games_df to match all_games_df platform names using the mapping method
mapping = {'N64':'Nintendo 64','PS':'PlayStation','PS3': 'PlayStation 3','DC':'Dreamcast','X360':'Xbox 360','XOne':'Xbox One',
          'Wii':'Wii','PC':'PC','PS2':'PlayStation 2','PS4':'PlayStation 4','GC':'GameCube','XB':'Xbox','WiiU':'Wii U',
          'GBA':'Game Boy Advance','3DS':'3DS','2600':'Atari','DS':'DS','PSV':'PlayStation Vita','PSP':'PSP','GB':'Game Boy',
           'NES':'Nintendo','PSP':'PlayStation Portable'}
video_games_df['platform'] = video_games_df['platform'].map(mapping)
video_games_df
```

```python
# Combine the data into a single dataset based on name of game and platform
merged_video_games_df = pd.merge(video_games_df, all_games_df, how='left', on=['name'])
merged_video_games_df.shape
```

```python
# Change the year_of_release column to the int type and verify column types
merged_video_games_df.loc[:, "year_of_release"] = merged_video_games_df.loc[:, "year_of_release"].astype("int")
merged_video_games_df.dtypes
```

```python
merged_video_games_df['summary'] = merged_video_games_df['summary'].str.replace(',', ' ')
merged_video_games_df['release_date'] = pd.to_datetime(merged_video_games_df['release_date'])
merged_video_games_df
```

```python
merged_video_games_df = merged_video_games_df.rename(columns={'platform_x': 'platform'})
merged_video_games_df
```
