```python
platform_agg_df = video_games_df.copy()
platform_agg_df.dropna()
platform_agg_df
```

```python
platform_agg_df = platform_agg_df.dropna()
```

```python
platform_agg_df["user_score"] = platform_agg_df["user_score"].astype(float)
```

```python
platform_agg_df
```

```python
aggregate_functions_by_name = {
    "year_of_release": "min",
    "developer": "first",
    "na_sales": "sum",
    "eu_sales": "sum",
    "jp_sales": "sum",
    "worldwide_sales": "sum",
    "critic_score": "mean",
    "user_score": "mean",
    "critic_count": "sum",
    "user_count": "sum",
}

platform_agg_df = platform_agg_df.groupby("name").agg(aggregate_functions_by_name)
platform_agg_df
```
